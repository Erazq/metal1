```python
# DÃ©finir les variables nÃ©cessaires
def calcul_effort_normal(f_y, A, gamma_M1, f_u, A_b, A_trous, gamma_M2):
    """
    Calcule les valeurs de N_Rd pour les deux cas:
    - Cas de la section brute
    - Cas de la section nette (avec des trous)

    :param f_y: Limite d'Ã©lasticitÃ© de l'acier (en MPa ou N/mmÂ²)
    :param A: Section brute (en mmÂ²)
    :param gamma_M1: Facteur de sÃ©curitÃ© pour la section brute
    :param f_u: Limite de rupture de l'acier (en MPa ou N/mmÂ²)
    :param A_b: Section brute (en mmÂ²)
    :param A_trous: Section des trous (en mmÂ²)
    :param gamma_M2: Facteur de sÃ©curitÃ© pour la section nette
    :return: Tuple contenant les deux valeurs de N_Rd (brute, nette)
    """
    # Cas de la section brute
    N_Rd_brute = (f_y * A) / gamma_M1

    # Cas de la section nette (avec des trous)
    A_net = A_b - A_trous
    N_Rd_nette = (0.9 * f_u * A_net) / gamma_M2

    return N_Rd_brute, N_Rd_nette

# Exemple d'utilisation
# DÃ©finir les paramÃ¨tres
f_y = 355  # MPa (limite d'Ã©lasticitÃ©)
A = 2000  # mmÂ² (section brute)
gamma_M1 = 1.05  # Facteur de sÃ©curitÃ© pour la section brute
f_u = 490  # MPa (limite de rupture)
A_b = 2000  # mmÂ² (section brute)
A_trous = 200  # mmÂ² (section des trous)
gamma_M2 = 1.25  # Facteur de sÃ©curitÃ© pour la section nette

# Calculer les efforts normaux
N_Rd_brute, N_Rd_nette = calcul_effort_normal(f_y, A, gamma_M1, f_u, A_b, A_trous, gamma_M2)

# Afficher les rÃ©sultats
print(f"Effort normal pour la section brute: {N_Rd_brute:.2f} kN")
print(f"Effort normal pour la section nette: {N_Rd_nette:.2f} kN")
```
